# Opgave 2 State pattern

## Klasse PrintText
De klasse **PrintText** print tekst op de console, waarbij de tekst om en om in hoofdletters en kleine letters wordt afgedrukt.

Lege regels worden wel afgedrukt, maar deze worden genegeerd als het gaat om het wisselen tussen hoofdletters en kleine letters.

Code van klasse **PrintText**:
```java
public class PrintText {

    private int type = 1;

    public void printText(String[] lines) {
        for(String line : lines) {
            printLine(line);
        }
    }

    public void printLine(String line) {
        if (line.isBlank()) {
            System.out.println();
        } else {
            if (type == 1) {
                System.out.println(line.toLowerCase());
            }
            if (type == 2) {
                System.out.println(line.toUpperCase());
            }
            type++;
            if (type > 2) {
                type = 1;
            }
        }
    }
    
}
```
